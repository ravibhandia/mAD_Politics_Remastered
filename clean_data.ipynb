```python
polling_data = pd.read_csv('./data/primary_poll_2020.csv')
polling_data.head()
```

```python
polling_data.drop(columns=['race', 'modeldate', 'pct_estimate', 'timestamp'], inplace=True)
```

```python
select_poll_data = polling_data.loc[polling_data['candidate_name'].str.contains('|'.join(candidates.keys()), case=False, regex=True)]
select_poll_data.reset_index(inplace=True)
select_poll_data.drop(columns=['index'], inplace=True)
select_poll_data.rename(columns={'pct_trend_adjusted': 'result', 'contestdate': 'Contest Date'}, inplace=True)
select_poll_data['candidate_name'] = select_poll_data['candidate_name'].apply(standardizeCandidateName)
select_poll_data.dropna(inplace=True)
select_poll_data
```

```python
final_poll_entity = candidate_entity.merge(select_poll_data, how='inner', on='candidate_name').merge(states, how='inner', left_on='state', right_on='state_name')
final_poll_entity = final_poll_entity[['state_id', 'candidate_id', 'Contest Date', 'result']]
final_poll_entity = final_poll_entity.groupby(['state_id', 'candidate_id', 'Contest Date'], as_index=False).agg({'result':max})
final_poll_entity.rename(columns={'Contest Date': 'contest_date'}, inplace=True)
final_poll_entity.index.name = 'poll_id'
final_poll_entity.head()
```
